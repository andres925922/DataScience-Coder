```python
# Seteamos la primera columna como índice
BA_SA_AIRBNB_DS = BA_SA_AIRBNB_DS.set_index("id")
columnas = list(BA_SA_AIRBNB_DS.columns)
BA_SA_AIRBNB_DS.head(6)
```

```python
# Listar valores nulos
zero_values = (BA_SA_AIRBNB_DS[columnas] == 0).sum()
print(zero_values)
```

```python
# listar valores nan
nan_values = (BA_SA_AIRBNB_DS[columnas] == np.nan).sum()
print(nan_values)
```

```python
# Obtenemos los varios que contienen mayor cantidad de alojamientos y graficamos
barrios_con_mas_alojamientos = BA_SA_AIRBNB_DS.groupby(["neighbourhood"])["name"].count().sort_values(ascending=False)
barrios_con_mas_alojamientos
barrios_con_mas_alojamientos.plot(
kind = "bar",
figsize = (10, 10),
xlabel = "Alojamientos",
ylabel = "Cantidad",
title = "Cantidad de alojamientos AIRBNB por barrio en Buenos Aires"
)
```

```python
# Obtenemos el precio promedio de los alquileres por zona
barrios_con_mas_alojamientos = BA_SA_AIRBNB_DS.groupby(["neighbourhood"])["price"].mean().sort_values(ascending=False)
barrios_con_mas_alojamientos
barrios_con_mas_alojamientos.plot(
kind = "bar",
figsize = (10, 10),
xlabel = "Alojamientos",
ylabel = "Precio promedio",
title = "Precio promedio de los alquileres AIRBNB por zona en Buenos Aires"
)
```

```python
# Gráfico de puntos con los precios de los alojamientos vs la cantidad de reviews.

BA_SA_AIRBNB_DS.plot.scatter(
y="price",
x="number_of_reviews",
c="blue",
s=6,
figsize = (10, 10),
)
```

```python
# Cantidad de alojamientos en función de los tipos de habitación, utilizando seaborn.

g = sbn.displot(
    data = BA_SA_AIRBNB_DS,
    x = "room_type", 
    aspect = 1,
    height = 6
)

g.set_axis_labels("Tipo de alojamiento", "Cantidad de alojamientos")
```

```python
# Number of reviews vs abailability per year

sbn.scatterplot(
    data = BA_SA_AIRBNB_DS,
    xlabel = "Numero de reviews",
    x = "number_of_reviews",
    y = "availability_365"
)
```

```python
# Price vs amount of reviews per month

sbn.scatterplot(
    data = BA_SA_AIRBNB_DS,
    x = "reviews_per_month",
    y = "minimum_nights",
    sizes = (20, 200),
    size = "price",
    hue = "room_type"
)
```
