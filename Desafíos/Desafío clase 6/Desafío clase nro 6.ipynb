3) Obtener un dataframe que contenga solo los días en que la acción bajó de precio y mostrar estos días con puntos rojos en una gráfica de tipo serie de tiempo.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
url = "https://raw.githubusercontent.com/JJTorresDS/stocks-ds-edu/main/stocks.csv"
df = pd.read_csv(url, index_col = 0)
print(df[["AMZN", "MCD", "SBUX", "GOOG", "MSFT"]].head(5).round(1))
print("============================================================")
print(df.describe())
print("============================================================")
print(df.info())
```

```python
# Graficamos inicialmente el comportamiento de una acción
df["AMZN"].plot(
    kind = "line",
    figsize = (10, 6),
    xlabel = "Date",
    ylabel = "Stock Price",
    title = "Price vs Date"
)
```

```python
# Obtener un dataframe donde se muestren los días en los que la acción bajó.
AMZN_series = df["AMZN"].pct_change()
"""
Percentage change between the current and a prior element.

Computes the percentage change from the immediately previous row by default. This is useful in comparing the percentage of change in a time series of elements.

"""

# Eliminamos los registros positivos y null
negative_changes_AMZN_series = AMZN_series.map(lambda x : x if (x < 0 and x is not np.nan) else None)
negative_changes_AMZN_series.dropna(inplace = True)

# Transformamos en un DataFrame
negative_changes_AMZN_df = pd.DataFrame({
    "date": negative_changes_AMZN_series.index,
    "stocks": negative_changes_AMZN_series.values
})

# negative_changes_AMZN_df = negative_changes_AMZN_df.set_index("date")

# Creamos el nuevo dataframe de stocks de AMAZON con la columna color en blue por definición
new_AMZ_df = pd.DataFrame({
    "date" : df["AMZN"].index,
    "stocks" : df["AMZN"].values,
    "color": "blue"
})

# new_AMZ_df = new_AMZ_df.set_index("date")
print(new_AMZ_df)

def agregar_color():
    """
    Función que permite cambiar el color en aquellas fechas de cambios negativos en función 
    de la existencia o no dentro del array de cambios negativos.
    """
    
    for i in new_AMZ_df["date"]:
        evaluate = i in list(negative_changes_AMZN_df["date"])
#         print(i, evaluate)
        if (evaluate):
#             juan = empleados.loc[empleados['Nombre'] == 'Juan']
            print(new_AMZ_df[new_AMZ_df["date"] == i]["color"])
            new_AMZ_df[new_AMZ_df["date"] == i]["color"] = "red"
            print(new_AMZ_df[new_AMZ_df["date"] == i]["color"])
            

            
agregar_color()

print(new_AMZ_df)


```

```python
# Graficamos las fechas donde los cambios fueron negativos
# negative_changes_AMZN_df.plot.scatter(x = "index", y = "stocks", s = "stocks", c = "red")

negative_changes_AMZN_df.plot(
    kind = "line",
    x = "index",
    figsize = (10, 6),
    xlabel = "Date",
    ylabel = "Stock change",
    title = "Negative change in stocks",
    marker = ".",
    markerfacecolor = "red",
    markersize = 12
)


```
